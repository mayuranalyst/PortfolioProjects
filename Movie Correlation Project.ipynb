```python
# In this project we are going to identfy correlations to answer the questions like
# which factors influence the movie's gross earnings
# Are there any other correlations that we can find in the data etc

#Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuation of plots we will create

# Read the data using pandas

df = pd.read_csv(r'C:\Users\mayur.LAPTOP-CQ8D0J84\Desktop\movies.csv')
```

```python
# Lets look at data first 10 rows

df.head(10)
```

```python
# Searching if there exists any missing data


for col in df.columns:
    prcnt_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,prcnt_missing))
    
```

```python
# Data types in our columns
df.dtypes
```

```python
#Change data type of columns from float64 to int64 for budget and gross columns
#But first we fill the missing cvalues with 0 for all numerical values

df['budget'] = df['budget'].fillna(0) 
df['gross'] = df['gross'].fillna(0)
#Now change the data type
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#sort movies by gross in descending order
df = df.sort_values(by = ['gross'], inplace = False, ascending = 0)
```

```python
#drop any duplicates to see distinct values for company
#However we dont save it in our data frame(because one company can produce multiple movies)

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
#We now find Correlations in data
#I am going to test my hypothesized correlations for example: High Budget can lead to a high gross film
#Or, company can have high correlation with Gross Earning(Influence of the brand)

#Scatter plot budget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget')

plt.show()

```

```python
#Lets do regression plot to confirm our hypothesis
#Plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
#Lets have a look at correlations in the data numerically
df.corr(method = 'pearson') # Types of correlations: Pearson , Kendall, Spearman
#Hence high correlation between budget and gross as seen from below table and above graph
```

```python
#Visualizing the correlation from the graph into a heatmap for easy comprehension 
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

#We see that Gross is highly influenced by budget and votes. Lets go further to confirm the same
```

```python
#Lets numerize the object fields in order to draw the conclusions on the non numeric firlds such as company name

#First lets copy our data frame so that we dont accidentally damage our main data frame
df_quantified = df

#Next, we convert the columns of data type 'object' into category
for col_name in df_quantified.columns:
    if(df_quantified[col_name].dtype == 'object'):
        df_quantified[col_name] = df_quantified[col_name].astype('category')
        df_quantified[col_name] = df_quantified[col_name].cat.codes
        
df_quantified        
```

```python
#Comparing below and above graphs we have confirm that categorizations were successful
# For example all the Genre fields which were "Action" are now categrorized as "0"
df
```

```python
# Now we apply same heatmap as earlier for the new quantified data frame

correlation_matrix = df_quantified.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix For New categorized Data Frame')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

#Higher the correlation, higher the coefficient seen in the heat map
 
```

```python
#Viewing the same in correlation matrix again
df_quantified.corr()
```

```python
#Lets unstack the above correlation matrix
correlation_matrix = df_quantified.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
#Lets sort correlation values and see the ones which have high correlation (Greater than 0.5)
sorted_pairs = corr_pairs.sort_values( ascending = False)

high_corr = sorted_pairs[(sorted_pairs) > 0.5 ]
high_corr

#Note: Same variable to same variable will always have high correlations we need to ignore that and see others
# Budget to Gross has highest correlation (0.75) 
# This is followed by Votes to Gross (0.63)
```

```python
# Hence, as seen from the graphs and other correlations above, we can conclude that Votes and Budget have highest correlations to the Gross Earnings of the film!!

# Company has low correlation to the Gross Earnings
```
